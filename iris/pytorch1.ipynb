```python
import os
import pandas as pd
from torchvision.io import read_image
```

```python
class DatasetClass(Dataset):
    def __init__(self,annonations_file, img_dir, transform=None, target_transform = None):
        self.img_labels = pd.read_csv(annonations_file)
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform
        
    def __len__(self):
        return len(self.img_labels)
    
    def __getitem__(self,idx):
        img_path = os.path.join(self.img_dir,self.img_labels.iloc[idx,0])
        image = read_image(img_path)
        lable = self.img_labels.iloc[idx,1]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            lable = self.target_transfrom(lable)
        return image,lable
    
        
        
        
        
```

```python
from torch import nn
```

Expected all tensors to be on the same device, but found at least two devices, cuda:0 and cpu! (when checking argument for argument mat1 in method wrapper_CUDA_addmm)Creatimg the neural network class

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28,512),
            nn.ReLU(),
            nn.Linear(512,512),
            nn.ReLU(),
            nn.Linear(512,10),
        )
    def forward(self,x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

```python
model = NeuralNetwork().to(device)
print(f"{model = }")
```

# Optimising the model

```python
from tqdm import tqdm
```

```python
def train_loop(dataloader:DataLoader, model:nn.Module, loss_fn:nn.CrossEntropyLoss, optimiser:torch.optim.Adam):
    size = len(dataloader.dataset)
    
    model.train()
    
    for batch,(X,y) in enumerate(dataloader):
        pred = model(X.to(device))
        loss = loss_fn(pred,y)
        
        loss.backward()
        optimiser.step()
        optimiser.zero_grad()
        
        if batch %100 == 0:
            loss,current = loss.item(),(batch+1)*len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
        
```

```python
def test_loop(dataloader,model,loss_fn):
    model.eval()
    size = len(dataloader.dataset)
    num_batches=len(dataloader)
    test_loss,correct = 0,0
    
    
    with torch.no_grad():
        for X,y in dataloader:
            pred = model(X.to(device))
            test_loss += loss_fn(pred,y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
            
```

```python
train_dataloader = DataLoader(traning_data , batch_size=64,shuffle=True,generator=torch.Generator(device=device))
test_dataloader = DataLoader(test_data, batch_size=64, shuffle=True,generator=torch.Generator(device=device))
```

```python
learning_rate = 1.5e-4
loss_fn = nn.CrossEntropyLoss()
optimiser = torch.optim.Adam(model.parameters(),lr=learning_rate)
epochs = 50
torch.set_default_device(device)
training_batch_size = len(train_dataloader)
validation_batch_size = len(test_dataloader)
epochs = tqdm(range(epochs), desc="Epochs")
for epoch in epochs:

    train_loop(dataloader=train_dataloader,model=model, loss_fn=loss_fn,optimiser=optimiser)
    test_loop(dataloader=test_dataloader,model=model,loss_fn=loss_fn)
print("done")
    
    
    
    
    
```
